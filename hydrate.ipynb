```python
isos = ["gb","de","it"]
for iso in isos:
    df1 = pd.read_csv('/home/atharva/NLP/Emotion_Analysis_Twitter/hydrated_tweets_{}.csv'.format(iso))
    f = open('/home/atharva/NLP/Emotion_Analysis_Twitter/Panacea/tweet_texts_{}.txt'.format(iso),"w")
    english_tweets = 0
    for i, text in enumerate(df1['full_text']):
        text = ' '.join(text.split('\n'))
        text = normalizeTweet(text)
        if df1['lang'][i] == 'en':
            f.write(text+'\n')
            english_tweets += 1
        if i%10000 == 0:
            print(i)
    f.close()
    print(english_tweets)
```
